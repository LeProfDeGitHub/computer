# Exercise 4
Express the matrix A as a linear combination of matrices representing projectors.

$$A = \begin{pmatrix}
 1 && 2\\
 2 && 1
\end{pmatrix}$$

**Answer:**

Projectors are matrices that satisfy
$P^2 = P$, and their eigenvalues are either 0 or 1.

$$A = 3\begin{pmatrix}
 0.5 && 0.5\\
 0.5 && 0.5
\end{pmatrix} - 1 \begin{pmatrix}
 0.5 && -0.5\\
 -0.5 && 0.5
\end{pmatrix}$$

Here is the numerical resolution:

```python
A = np.array([
    [ 1, 2],
    [2,  1],
    ])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Construct the projectors
projectors = []
for i in range(len(eigenvalues)):
    projector = np.outer(eigenvectors[:, i], eigenvectors[:, i].conj())
    projectors.append(pd.DataFrame(projector))

# Express A as a linear combination of projectors
A_as_linear_combination = sum([eigenvalues[i] * projectors[i] for i in range(len(eigenvalues))])

print("Matrix A as a linear combination of projectors:")
print(f"A =\n {A_as_linear_combination}\n  = {eigenvalues[0]:.0f} x\n {projectors[0]}\n  +\n {eigenvalues[1]:.0f} x {projectors[1]}")
```
